```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import CoolProp as CP
from CoolProp.CoolProp import PropsSI

# Check: CoolProp version
print( CP.__version__)
print(CP.__gitrevision__)

# Constants
golden = (1 + 5 ** 0.5) / 2
kilo = 1e3
Mega = 1e6
nPoints = 1000

# helper function: get slope using two sorted arrays
def numSlopeAr(xAr, yAr):
 deltaX = np.empty(len(xAr))
 deltaY = np.empty(len(yAr))
 slopeAr = np.empty(len(yAr))
 for index in range(1,len(xAr)-1):
     deltaX[index] = xAr[index-1] - xAr[index+1]
     deltaY[index] = yAr[index-1] - yAr[index+1]
     slopeAr[index] = deltaY[index]/deltaX[index]
 # inaccurate, but who cares?
 slopeAr[0]=slopeAr[1]
 slopeAr[-1]=slopeAr[-2]  
 return slopeAr

# helper function: draw tangent from array
def drawTangent(xAr, yAr, slopeAr, index: int):
 xVal = xAr[index]
 yVal = yAr[index]
 slopeVal = slopeAr[index]
 # line length as percentage of length of x axis
 xRange = abs(max(xAr)-min(xAr))
 xRangeAr = np.arange(xVal-0.15*xRange, xVal+0.15*xRange)
 tang = yVal + slopeVal*(xRangeAr-xVal)
 plt.plot(xVal,yVal,'om',xRangeAr,tang,'-k')
```

```python
# Get all the data
FluidName = 'Propane'
T_crt = PropsSI('TCRIT',FluidName)
T_trp = PropsSI('TTRIPLE',FluidName)
p_crt = PropsSI('PCRIT',FluidName)
p_trp = PropsSI('PTRIPLE',FluidName)
d_crt = PropsSI('RHOCRIT',FluidName)
d_trp_liq = PropsSI('D','T',T_trp,'Q',0,FluidName)
d_trp_vap  = PropsSI('D','T',T_trp,'Q',1,FluidName)

# Saturation data
Ts_sat = np.linspace(T_trp, T_crt, num=nPoints)
TcrtTsat = T_crt/Ts_sat
ps_sat = CP.CoolProp.PropsSI('P','T',Ts_sat,'Q',0,FluidName)
logpsat = np.log(ps_sat/p_crt)
ds_sat_liq = CP.CoolProp.PropsSI('D','T',Ts_sat,'Q',0,FluidName)
ds_sat_vap = CP.CoolProp.PropsSI('D','T',Ts_sat,'Q',1,FluidName)
ds_rec = (ds_sat_liq+ds_sat_vap)/2
vs_sat_liq = 1/ds_sat_liq
vs_sat_vap = 1/ds_sat_vap
ss_sat_liq = CP.CoolProp.PropsSI('S','T',Ts_sat,'Q',0,FluidName)
ss_sat_vap = CP.CoolProp.PropsSI('S','T',Ts_sat,'Q',1,FluidName)
hs_sat_liq = CP.CoolProp.PropsSI('H','T',Ts_sat,'Q',0,FluidName)
hs_sat_vap = CP.CoolProp.PropsSI('H','T',Ts_sat,'Q',1,FluidName)
us_sat_liq = CP.CoolProp.PropsSI('U','T',Ts_sat,'Q',0,FluidName)
us_sat_vap = CP.CoolProp.PropsSI('U','T',Ts_sat,'Q',1,FluidName)
```

```python
# Some general overview saturation plots
plt.figure(figsize=(16,16*2/3/golden))
plt.subplot(2,3,1)
plt.plot(Ts_sat, ps_sat/Mega, color='blue')
plt.xlabel('Temperature in K')
plt.ylabel('Pressure in MPa')
# plt.yscale('log')

plt.subplot(2,3,2)
plt.plot(ds_sat_liq, Ts_sat, color='blue')
plt.plot(ds_sat_vap, Ts_sat, color='red')
plt.plot(ds_rec, Ts_sat, color='black')
plt.xlabel('Density in kg/m³')
plt.ylabel('Temperature in K')

plt.subplot(2,3,3)
plt.plot(ds_sat_liq, ss_sat_liq/kilo, color='blue')
plt.plot(ds_sat_vap, ss_sat_vap/kilo, color='red')
plt.xlabel('Density in kg/m³')
plt.ylabel('Specific entropy in kJ/(kg·K)')

plt.subplot(2,3,4)
plt.plot(TcrtTsat, logpsat, color='blue')
plt.xlabel('Tc/Ts')
plt.ylabel('ln(ps/pc)')

plt.subplot(2,3,5)
plt.plot(hs_sat_liq/kilo, ss_sat_liq/kilo, color='blue')
plt.plot(hs_sat_vap/kilo, ss_sat_vap/kilo, color='red')
plt.xlabel('Specific enthalpy in kJ/kg')
plt.ylabel('Specific entropy in kJ/(kg·K)')

plt.subplot(2,3,6)
plt.plot(vs_sat_liq, us_sat_liq/kilo, color='blue')
plt.plot(vs_sat_vap, us_sat_vap/kilo, color='red')
plt.xscale('log')
plt.xlabel('Specific volume in m³/kg')
plt.ylabel('Specific energy in kJ/kg')
```

```python
# Clausius-Clapeyron-Array
dpdT = (ss_sat_vap-ss_sat_liq) / (vs_sat_vap-vs_sat_liq)
dTdp = 1/dpdT

# pick some point from sat line
mySatPoint = int(nPoints-150)

# replot ps=ps(Ts)
plt.plot(Ts_sat, ps_sat, color='blue')
plt.xlabel('Temperature in K')
plt.ylabel('Pressure in Pa')
# then put a marker on point and drawTangent(x,y,slope,pointIndex)
drawTangent(Ts_sat, ps_sat, dpdT, mySatPoint)
```

```python
# liq or vap side?
vapSide = False
if vapSide:
    ds_sat = ds_sat_vap
    vs_sat = vs_sat_vap
    ss_sat = ss_sat_vap
    hs_sat = hs_sat_vap
    us_sat = us_sat_vap
else:    
    ds_sat = ds_sat_liq
    vs_sat = vs_sat_liq
    ss_sat = ss_sat_liq
    hs_sat = hs_sat_liq
    us_sat = us_sat_liq
```

```python
# replot plot of interest
plt.plot(hs_sat_liq, ss_sat_liq, color='blue')
plt.plot(hs_sat_vap, ss_sat_vap, color='red')
plt.xlabel('Specific enthalpy in J/kg')
plt.ylabel('Specific entropy in J/(kg·K)')

# get numerical derivative
numSlope = numSlopeAr(hs_sat,ss_sat)

# calculate analytical first derivative ds/dh@sigma
ds_dT_p = CP.CoolProp.PropsSI('d(Smass)/d(T)|P','D',ds_sat,'T',Ts_sat,FluidName)
ds_dp_T = CP.CoolProp.PropsSI('d(Smass)/d(P)|T','D',ds_sat,'T',Ts_sat,FluidName)
ds_dT_sigma = ds_dT_p+ds_dp_T*dpdT
dh_dT_p = CP.CoolProp.PropsSI('d(Hmass)/d(T)|P','D',ds_sat,'T',Ts_sat,FluidName)
dh_dp_T = CP.CoolProp.PropsSI('d(Hmass)/d(P)|T','D',ds_sat,'T',Ts_sat,FluidName)
dh_dT_sigma = dh_dT_p+dh_dp_T*dpdT

ds_dh_sigma = ds_dT_sigma/dh_dT_sigma
anaSlope = ds_dh_sigma

# pick some point (vap peak between 86 and 87)
mySatPoint = int(nPoints-870)
# print values of that point and draw tangent
print("num slope = " + str(numSlope[mySatPoint]))
print("ana slope = " + str(anaSlope[mySatPoint]))
print("avg num-ana difference = " + str(np.nanmean(numSlope-anaSlope)))
print("max num-ana difference = " + str(max(numSlope-anaSlope)))
drawTangent(hs_sat, ss_sat, numSlope, mySatPoint)
```

```python
# plot anaSlope over hs_sat
plt.plot(hs_sat, anaSlope, color='cyan')
plt.xlabel('Specific enthalpy in J/kg')

# get numerical 2nd deriv
numSlope2nd = numSlopeAr(hs_sat,anaSlope)

# calculate analytical 2nd deriv d2s_dh2

# draw tangent
drawTangent(hs_sat, anaSlope, numSlope2nd, mySatPoint)
```
